# 5. Limpieza del dataset

```python
df_limpio = df.copy()

# Limpiar nombres de columnas
df_limpio.columns = [columna.strip().replace(" ", "_").replace("-", "_")
   for columna in df_limpio.columns]

# Limpiar espacios en columnas de texto
for columna in df_limpio.select_dtypes(include="object").columns:
    df_limpio[columna] = df_limpio[columna].str.strip()

# Eliminar duplicados
duplicados = df_limpio.duplicated().sum()
df_limpio = df_limpio.drop_duplicates().reset_index(drop=True)

print("Duplicados eliminados:", duplicados)
print("Tamaño final:", df_limpio.shape)

df_limpio.head()
```

# 6. Crear variables nuevas

Como el dataset original carece de una columna fecha, se crea una fecha ficticia en función de la secuencia de los registros para efectos de validar la temporalidad, además de otras varialbles útiles para el análisis final de la información.

```python
df_limpio["Fecha_Estimada"] = pd.to_datetime("2014-01-01") + pd.to_timedelta(df_limpio.index // 10,unit="D")

df_limpio["Anio"] = df_limpio["Fecha_Estimada"].dt.year
df_limpio["Mes"] = df_limpio["Fecha_Estimada"].dt.month
df_limpio["Dia_Semana"] = df_limpio["Fecha_Estimada"].dt.day_name()

df_limpio["Margen_Profit"] = np.where(df_limpio["Sales"] != 0,df_limpio["Profit"] / df_limpio["Sales"],0)

df_limpio["Venta_Promedio_Unidad"] = np.where(df_limpio["Quantity"] != 0,df_limpio["Sales"] / df_limpio["Quantity"],0)

df_limpio.head()
```

#   7. Guardar dataset limpio

```python
df_limpio.to_csv("dataset/SampleSuperstore_limpio.csv", index=False)
print("Dataset limpio guardado correctamente.")
```
